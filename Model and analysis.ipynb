# Random Forest Regression for predicting and ranking importance of various features 

## Importing the packages and data

```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
dataset = pd.read_csv('cosesamp.csv')
```

```python
dataset.head()
```

```python
y = dataset['total_cases_per_million']
X = dataset.iloc[:, 7:]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=1000, max_features=4, max_depth=7,  random_state=18, max_samples=0.5)
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)
```

```python
imp = regressor.feature_importances_
imp_percent = []
for i in imp:
    imp_percent.append(i*100)
print(imp_percent)
imp_df = np.array(imp_percent)
```

```python
from sklearn import metrics
np.sqrt(metrics.mean_squared_error(y_pred, y_test))
```
